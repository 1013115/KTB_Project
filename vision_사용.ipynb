```python
import pyheif
from PIL import Image

def convert_heic_to_jpg(input_path, output_path):
    # HEIC 파일 읽기
    heif_file = pyheif.read(input_path)

    # HEIC 데이터를 PIL 이미지로 변환
    image = Image.frombytes(
        heif_file.mode,
        heif_file.size,
        heif_file.data,
        "raw",
        heif_file.mode,
        heif_file.stride,
    )

    # JPG로 저장
    image.save(output_path, "JPEG")
    print(f"Image saved as {output_path}")

def convert_heic_to_png(input_path, output_path):
    # HEIC 파일 읽기
    heif_file = pyheif.read(input_path)

    # HEIC 데이터를 PIL 이미지로 변환
    image = Image.frombytes(
        heif_file.mode,
        heif_file.size,
        heif_file.data,
        "raw",
        heif_file.mode,
        heif_file.stride,
    )

    # PNG로 저장
    image.save(output_path, "PNG")
    print(f"Image saved as {output_path}")

# HEIC 파일을 JPG 또는 PNG로 변환하는 예시
heic_file_path = "/content/IMG_1856.HEIC"
jpg_output_path = "output.jpg"
png_output_path = "output.png"

# HEIC -> JPG 변환
convert_heic_to_jpg(heic_file_path, jpg_output_path)

# HEIC -> PNG 변환
convert_heic_to_png(heic_file_path, png_output_path)
```

사진을 따로 저장함 -> 이건 나중에 어디에 저장하는건가? 아님 날라갈건지


```python
import base64
from PIL import Image
import pyheif
import os
from openai import OpenAI

# Function to convert HEIC to PNG or JPG using pyheif
def convert_heic_to_jpg_or_png(heic_image_path, output_format='PNG'):
    # Read the HEIC image using pyheif
    heif_file = pyheif.read(heic_image_path)

    # Convert to Pillow Image
    image = Image.frombytes(
        heif_file.mode,
        heif_file.size,
        heif_file.data,
        "raw",
        heif_file.mode,
        heif_file.stride,
    )

    # Save the image in the desired format (PNG or JPG)
    new_image_path = os.path.splitext(heic_image_path)[0] + "." + output_format.lower()
    image.save(new_image_path, output_format)

    return new_image_path

# Function to convert and save the image as PNG or JPG
def convert_to_jpg_or_png(image_path):
    valid_extensions = ['.png', '.jpg', '.jpeg']
    file_name, ext = os.path.splitext(image_path)

    # If the image is already in PNG or JPG format, return the same path
    if ext.lower() in valid_extensions:
        return image_path

    # If the file is HEIC, convert it using pyheif
    if ext.lower() == '.heic':
        return convert_heic_to_jpg_or_png(image_path)

    # Otherwise, attempt to open and convert it using Pillow
    image = Image.open(image_path)
    new_image_path = file_name + ".png"
    image.save(new_image_path, "PNG")
    return new_image_path

# Function to encode the image
def encode_image(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode('utf-8')



client = OpenAI()

# Path to your image (you can test with various formats like .heic, .bmp, etc.)
image_path = "/content/IMG_1890.PNG"

# Convert the image if necessary
converted_image_path = convert_to_jpg_or_png(image_path)

user_input = "paper receipt"

# Get the base64 string of the converted image
base64_image = encode_image(converted_image_path)

# Creating the request to OpenAI API
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {
            "role": "user",
            "content": [
                {
                    "type": "text",
                    "text": "what's in this image. Speak only objects that are clearly visible.",
                },
                {
                    "type": "image_url",
                    "image_url": {
                        "url":  f"data:image/jpeg;base64,{base64_image}"
                    },
                },
            ],
        },
        {
            "role":"user",
            "content":f"Does this image have a {user_input}? Answer Yes or No."
        }
    ],
    max_tokens=300,
)

print(response.choices[0].message)
```

## chain 사용


```python
import os
import base64
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, SystemMessage
from langchain_core.runnables import (
    RunnablePassthrough,
    RunnableLambda,
    RunnableParallel,
    chain,
)

# Vision 모델을 사용하기 위한 ChatOpenAI 클래스의 인스턴스를 생성합니다.
vision_model = ChatOpenAI(
    model="gpt-4o-mini", temperature=0.1, max_tokens=8196
)

# 이미지 파일을 base64로 인코딩하는 함수입니다.


def encode_image(image_path):
    with open(image_path, "rb") as image_file:
        img = base64.b64encode(image_file.read()).decode("utf-8")

    return f"data:image/jpeg;base64,{img}"


# 이미지 파일의 경로와 이미지 파일의 URL을 입력으로 받아서 처리하는 파이프라인입니다.
@chain
def vision_chain(inputs):
    image_path, url, user_input = str(inputs["image_path"]), bool(inputs["url"]), str(inputs["user_input"])
    vision_model = ChatOpenAI(model="gpt-4o-mini", max_tokens=256)
    if url:
        image_url = image_path
    else:
        base64_image = encode_image(image_path)
        image_url = f"{base64_image}"

    messages = [
        {"role":"assistant", "content": "I am an image analyst."},
        {"role": "user",
         "content": [
          {
            "type": "text",
            "text": "what's in this image. Speak only objects that are clearly visible.",
          },
          ],
        },
        {"role": "user",
         "content": [
          {
            "type": "text",
            "text": f"Does this image have a {user_input}? Answer Yes or No",
          },
          {
             "type": "image_url",
                "image_url": {
                    "url": image_url,
                    "detail": "auto",
                },
          },
        ],
      },

    ]
    output = vision_model.invoke(messages)
    return output.content
```

```python
final_chain = vision_chain | StrOutputParser()
```

```python
image_path = "/content/IMG_1890.PNG"
final_chain.invoke({"image_path": image_path, "url": False, "user_input":"tumbler"})
```
